### 1. Write a simple program to compute all the roots of a quadratic equation.

**Hint formulas**

$$
r = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

$$
\text{Real and distinct roots if } b^2 - 4ac > 0
$$

$$
\text{Real and equal roots if } b^2 - 4ac = 0
$$

$$
\text{Complex conjugate roots if } b^2 - 4ac < 0
$$



```python
def compute_roots(a, b, c):
    D = b**2 - 4*a*c # Discriminant

    _r1, _r2 = 0,0

    match D:
        case 0:
            _r1 = (-1*b)/(2*a)
            _r2 = _r1
        case _: # default case
            if D>0:
                _r1 = ((-1*b)+D**0.5)/(2*a)
                _r2 = ((-1*b)-D**0.5)/(2*a)
            elif D<0:
                return "Imaginary roots"
        
    return f"root 1: {round(_r1,2)} and root2: {round(_r2,2)}"
        
```

```python
compute_roots(5,14,5)
```

### 3. Find the largest among three numbers.

```python
def largest_among_three_numbers(a, b, c):
    if a>b:
        if a>c:
            return a
        else:
            return c
    else:
        if b>c:
            return b
        else:
            return c
```

```python
largest_among_three_numbers(10, 12, 14)
```

### 4. Illustrate the `LEFT SHIFT` and `RIGHT SHIFT` operations using operators. 

```python
def shift_operation(n, x):
    _choice = input("Type 'L' for left shift, 'R' for right shift operation...")

    match _choice:
        case 'L':
            n = n<<x
        case 'R':
            n = n>>x
        case _: # default case
            return "Invalid choice. Try again."
        
    return f"Final value after {_choice} operation: {n}"
        
```

```python
shift_operation(11, 3)
```

``11`` in binary: ``01011``
``11`` after ``3`` shifts to the left: ``01011000`` -> `88`

### 5. Check whether the given number is zero, positive or negative using `if-elif-else` ladder.

```python
def positive_negative_zero(n):
    if n==0:
        return "Zero."
    elif n>0:
        return "Positive"
    else:
        return "Negative"
```

```python
positive_negative_zero(-12)
```

### 6. Find the smallest among three numbers using `conditional` operator and `lambda` expression.

```python
find_smallest = lambda a,b,c: a if (a<b and a<c) else (b if (b<a and b<c) else c)
```

```python
find_smallest(10,12,14)
```

### 7. Accept the number of days a member is late to return the book. Calculate and display the fine with the appropriate message using `if-elif-else` ladder. The fine is charged as per the table below:

| Late Period | Fine |
|-------------|------|
| 5 days | € 0.50 |
| 6 - 10 days | € 1.00 |
| 10+ days | € 5.00 |
| 30+ days | € 10.00 |

```python
def library_fine(days):
    d = days
    _fine = 0.0

    if d==5:
        _fine = d * 0.5
    elif d>5 and d<=10:
        _fine = (5 * 0.5) + ((d-5) * 1.0)
    elif d>10 and d<30:
        _fine = (5 * 0.5) + (5 * 1.0) + ((d-10) * 5.0)
    elif d>30:
        _fine = (5 * 0.5) + (5 * 1.0) + (20 * 5.0) + ((d-30) * 10.0)

    return f"You have to pay € {_fine}."
```
