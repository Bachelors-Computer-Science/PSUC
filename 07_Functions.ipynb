### 1. `FACTORIAL`: Write a function `fact()` to find the factorial of a given number. Using this function, compute ``nCr`` function.

$$
^nC_r = \binom{n}{r} = \frac{n!}{r!(n-r)!}
$$


```python
from typing import List
```

```python
def fact(n):
    _prod = 1
    for i in range(2,n+1):
        _prod *= i

    return _prod
```

```python
def fact_rec(n):
    if n==0 or n==1:
        return 1
    
    return n * fact_rec(n-1)
```

```python
fact_rec(5)
```

```python
def combination(n, r):

    # Formula: nCr
    ## fact(n) / ( fact(r) * fact(n-r))
     
    return fact(n) / ( fact(r) * fact(n-r)) 
```

```python
combination(10,2)
```

### 2. `PRIME`: Write a function `is_prime()` to check whether the given number is prime or not. Using this function, generate first `N` prime numbers.

```python
def is_prime(n):
    if n<=1:
        return False
    
    for i in range(2,n//2+1):
        if n%i==0:
            return False
        
    return True
```

```python
def generate_first_N_primes(n):
    number = 2
    count = 0
    while(count<n):
        if(is_prime(number)):
            print(f"{count+1}. {number}")
            count += 1
        number += 1
```

```python
generate_first_N_primes(10)
```

### 3. Write a function `largest()` to find the maximum of a given list of numbers.

```python
def largest(arr:List):
    n = len(arr)
    if n==1:
        return arr[0]
    
    _max = arr[0]
    for i in range(1,n):
        _max = max(_max, arr[i])
    
    return _max
```

```python
largest([123,456,789,135,246,579,867,532,431])
```

### 4. `Corner Function Matrix`: Write a function `corner_sum()` which takes as parameter: number of rows , number of columns of a matrix and the matrix itself and returns the sum of the elements in the four corners of the matrix.

```python
def corner_sum(m, n, A:List[List]):
    return A[0][0] + A[0][n-1] + A[m-1][0] + A[m-1][n-1]
```

```python
corner_sum(3,3,
    [
        [1, 2, 3],
        [4, 5, 6],
        [7, 8, 9]
    ]
)
```

### 5. `GCD`: Write a recursive function `gcd()` to find the GCD of two numbers. 

```python
def gcd(a, b):
    if b > a:
        return gcd(b, a)
    
    if b==0:
        return a
    
    return gcd(b, a%b) 
```

```python
gcd(7, 3)
```

### 6. `FIBONACCI`: Write a recursive function `fib()` to generate ``n``th Fibonacci term. Then generate first `N` Fibonacci terms using this function.

```python
def fib(n):
    if n==0 or n==1:
        return n
    
    return fib(n-1) + fib(n-2)
```

```python
n = int(input("Enter the value of n: "))
for i in range(n):
    print(fib(i))
```

### 7. `PRINT`: Write a recursive function `print_rec()` to print `1` to `N` numbers using recusion. Do not use loops inside the function.

```python
def print_rec(n):
    if n==1:
        print(f"{n}")
        return
    print_rec(n-1)
    print(n)
```

```python
print_rec(5)
```

### 8. `MULTIPLICATION`: Write a recursive function to multiply two numbers.

```python
def multiply(a, b):
    if b==1:
        return a
    
    return a+multiply(a, b-1)
```

```python
multiply(23,45)
```

### 9. `Tower of Hanoi`: Simulate the working of Tower of Hanoi for `n` disks. Print the number of moves.

```python
def toh(n, source, interim, destination, moves):

    # source: A
    # interim: B
    # destination: C

    if (n==1):
        print(f"Move disk 1 from {source} to {destination}")
        return 1
    
    x = toh(n-1, source, destination, interim, moves+1) # A, C, B

    print(f"Move disk {n} from {source} to {destination}")

    y = toh(n-1, interim, source, destination, moves+1) # B, A, C

    return x + 1 + y
```

```python
moves = toh(4, 'A', 'B', 'C', moves=0)
```
