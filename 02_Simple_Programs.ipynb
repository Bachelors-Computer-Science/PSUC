### 1. Write a program to compute simple and compound interest.

**Hint formulas**

Simple interest: $$ SI = \frac{PNR}{100} $$

Compound interest (amount of interest only):

$$
CI = P\left(1 + \frac{R}{100}\right)^N - P
$$

where \(P\) = principal, \(N\) = time period, \(R\) = rate of interest.

```python
def simple_interest(principal, time_period, rate_of_interest):
    _p = principal
    _n = time_period
    _roi = rate_of_interest

    S_I = (_p * _n * _roi) / 100

    return S_I
```

```python
def compound_interest(principal, time_period, rate_of_interest):
    _p = principal
    _n = time_period
    _roi = rate_of_interest

    C_I = _p * ((1 + _roi/100)**_n) - _p

    return C_I
```

```python
simple_interest(100, 3, 5)
```

```python
compound_interest(100, 3, 5)
```

### 2. Input radius to find the volume and surface area of a sphere.

**Hint formulas**

Volume:

$$
\text{Volume} = \frac{4\pi r^3}{3}
$$

Surface area:

$$
\text{Area} = 4\pi r^2
$$

where \(r\) is the radius of the sphere.

```python
def sphere_volume(radius):
    import math

    _r = radius
    _pi = math.pi

    V = (4 * _pi * (pow(_r,3)))/3

    return V
```

```python
def sphere_area(radius):
    import math

    _r = radius
    _pi = math.pi

    A = (4 * _pi * (pow(_r,2)))

    return A
```

```python
sphere_volume(10)
```

```python
sphere_area(10)
```

### 3. Convert the given temperature in Fahrenheit to Centigrade.

**Hint formula**

$$
C = \frac{5}{9}(F - 32)
$$

where \(F\) is Fahrenheit and \(C\) is Centigrade.

```python
def fahrenheit_to_centigrade(F):
    C = (5/9)*(F - 32)
    return C
```

```python
fahrenheit_to_centigrade(-40)
```

```python
fahrenheit_to_centigrade(98.6)
```

```python
def return_stoopid_maths(a, b):
    _sum = a + b
    _diff = a - b
    _prod = a * b
    try:
        _div = a / b
    except:
        print("ZeroDivisionError")

    print(f"Sum: {_sum}")
    print(f"Diff: {_diff}")
    print(f"Prod: {_prod}")
    print(f"Division: {_div}")
```
