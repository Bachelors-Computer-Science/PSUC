## Problems

### 1. Write a program to find the area of a circle.

```python
def area_of_the_circle(radius):
    import math

    _r = radius
    _pi = math.pi
    return _pi * (_r ** 2) 
```

```python
area_of_the_circle(10)
```

### 2. Write a program to find the largest of 3 numbers.

```python
def largest_of_3_numbers(a, b, c):
    return (a
            if (a > b and a > c)
            else (b
                  if (b > a and b > c)
                  else c
            )
    )
```

```python
largest_of_3_numbers(10, 12, 14)
```

### 3. Write a program to find the factorial of a number.

```python
def factorial(n):
    
    _product = 1
    for i in range(1,n+1):
        _product *= i

    return _product
```

```python
factorial(5)
```
