### 1. Count the number of words in a sentence.

```python
from typing import List
```

```python
def count_words(sentence:str)->int:
    sentence_split = sentence.split(" ")
    return len(sentence_split)
```

```python
def count_words_2(sentence:str)->int:
    _count = 0
    n = len(sentence)

    for i in range(n):
        if i==n-1:
            _count+=1
        elif i<n-1:
            if sentence[i]==" " and sentence[i+1]!=" ":
                _count+=1

    return _count
```

```python
count_words_2("I.")
```

```python
count_words_2("I")
```

```python
count_words_2("I am Ryan Literally Me Gosling.")
```

### 2. Toggle the case of every character in a string.

```python
def toggle_case(_str:str):
    _str_toggle = ""

    for c in _str:
        if c.islower():
            _str_toggle += c.upper()
        else:
            _str_toggle += c.lower()

    return _str_toggle
```

```python
toggle_case("I aM ryAn GosLInG.")
```

### 4. Arrange `n` names in alphabetical order.

```python
def bubble_sort(str_arr:List[str]):
    n = len(str_arr)
    arr = str_arr.copy()

    for i in range(n):
        swapped_flag = False
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped_flag = True

        if not swapped_flag:
            break

    return arr 
```

```python
def sort_n_names_manual(_str_list:List[str]):
    return bubble_sort(_str_list)
```

```python
def sort_n_names(_str_list:List[str]):
    return sorted(_str_list)
```

```python
sort_n_names_manual(["Ryan", "Gosling", "ryan", "gosling", "Blade", "runner"])
```

```python
sort_n_names(["Ryan", "Gosling", "ryan", "gosling", "Blade", "runner"])
```

### 5. Delete a word from a given sentence.

```python
def delete_word_from_sentence(sentence:str, _word:str):
    import string

    word_list = sentence.split()
    # print(word_list)
    result_sentence = []

    for word in word_list:
        clean_word = word.strip(string.punctuation) # to remove delimitters

        if clean_word.lower() != _word.lower():
            result_sentence.append(word)
        
    return " ".join(result_sentence)
```

```python
sentence = "I, am Ryan Gosling; he is LITERALLY me !"
word = "Gosling"

delete_word_from_sentence(sentence, word)
```

### 6. Search for a substring in a given string.

```python
def search_substr(string:str, substr:str):
    str_len = len(string)
    substr_len = len(substr)

    if substr_len > str_len:
        return -1 # substring cannot be longer than string
    
    remaining_len = str_len - substr_len
    for start in range(remaining_len + 1):
        match_flag = True
        for end in range(substr_len):
            if string[start+end] != substr[end]:
                match_flag = False
                break
            
        if match_flag == True:
            return start
        
    return -1
```

```python
search_substr("Ryan Gosling is Literally Me", "Gosling")
```

### 7. Delete all repeated words in a given string.

```python
def delete_repeated_words(sentence:str):
    import string

    word_list = sentence.split()
    seen_set = []
    result_set = []

    for word in word_list:
        clean_word = word.strip(string.punctuation)

        if clean_word not in seen_set:
            seen_set.append(clean_word)
            result_set.append(word)

    return " ".join(result_set)

delete_repeated_words("I am Ryan Gosling, Ryan from Blade Runner 2024 Gosling.") # not working for the word "Gosling"
```
